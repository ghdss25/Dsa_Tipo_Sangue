## Importação de Bibliotecas para a Planilha

```python
import pandas as pd 
import numpy as np 
import sqlite3 
from datetime import date, datetime
```

## Importação da Planilha de Tipo Sanguíneo do Usuário 

```python
tipo_sangue = pd.read_excel('tipo_sangue.xlsx')

tipo_sangue
```

```python
## Conversão da Ultima Consulta em formato dia, mês e ano 

tipo_sangue['Última Consulta'] = pd.to_datetime(tipo_sangue["Última Consulta"]) 
tipo_sangue['Ano_Consulta'] = tipo_sangue['Última Consulta'].dt.year
tipo_sangue['Mês_Consulta'] = tipo_sangue['Última Consulta'].dt.month
tipo_sangue['Dia_Consulta'] = tipo_sangue['Última Consulta'].dt.day
```

```python
display(tipo_sangue)
```

## Cálculo do Imc dos Usuários

```python
tipo_sangue['Imc'] = (tipo_sangue['Peso'] / (tipo_sangue['Altura'] * tipo_sangue['Altura'])) 
tipo_sangue
```

## Apagar as colunas de Peso e Altura 

```python
tipo_sangue = tipo_sangue.drop(["Peso", "Altura"], axis=1)
tipo_sangue
```

```python
## Transformar o Imc em duas casas decimais 
tipo_sangue['Imc'] = tipo_sangue['Imc'].round(2)
tipo_sangue
```

## Mesclando Nome e Sobrenome

```python
tipo_sangue['Nome_Completo'] = tipo_sangue['Nome'] + '  ' + tipo_sangue['Sobrenome']
```

```python
tipo_sangue
```

## Apagando Nome e o Sobrenome dos usuários

```python
tipo_sangue = tipo_sangue.drop(["Nome", "Sobrenome"], axis=1)
tipo_sangue
```

## Conexão para  o Banco de dados 

```python
dbconn = sqlite3.connect('tipo_sangue.db')
```

```python
print(dbconn)
```

# Definição de Star Schema para Dim_Usuario

```python
dbconn.execute('''
    CREATE TABLE Dim_Usuario (
        id INTEGER PRIMARY KEY,
        "Nome_Completo" TEXT,
        "Título" TEXT,
        "E-mail" TEXT,
        "Sexo" TEXT,
        "Profissão" TEXT,
        "Nascimento" TEXT
    )
''')

Dim_Usuario.to_sql('Dim_Usuario', dbconn, if_exists='append', index=False)
```

## Definição de Star Schema para Dim_Telefone

```python
# Define the schema for Dim_Telefone
dbconn.execute('''
    CREATE TABLE Dim_Telefone (
        id INTEGER PRIMARY KEY,
        "Telefone" TEXT
    )
''')

Dim_Telefone.to_sql('Dim_Telefone', dbconn, if_exists='append', index=False)
```

# Definição de Star Schema para Dim_Situacao_Sanguinea

```python
dbconn.execute('''
    CREATE TABLE Dim_Situacao_Sanguinea (
        id INTEGER PRIMARY KEY,
        "Tipo_Sanguineo" TEXT,
        "Última_Consulta" DATE,
        "Ano_Consulta" TEXT, 
        "Mês_Consulta" TEXT, 
        "Dia_Consulta" TEXT,
        "Imc" REAL
    )
''')

Dim_Situacao_Sanguinea.to_sql('Dim_Situacao_Sanguinea', dbconn, if_exists='append', index=False)
```

## Definição de Star Schema para Dim_Endereco

```python
dbconn.execute('''
    CREATE TABLE Dim_Endereco (
        id INTEGER PRIMARY KEY,
        "Código_Postal" TEXT,
        "Cidade" TEXT,
        "Endereço" TEXT
    )
''')

Dim_Endereco.to_sql('Dim_Endereco', dbconn, if_exists='append', index=False)
```

## 1 - Total os Imcs por Títulos

```python
query = '''
    
    SELECT Título, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média FROM Dim_Usuario INNER JOIN Dim_Situacao_Sanguinea 
    ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id GROUP BY Título
    
'''

lista_imc_usuario = pd.read_sql_query(query, con=dbconn)
lista_imc_usuario
```

## 2 - Total os Imcs por Profissão

```python
query1 = '''
    
    SELECT Profissão, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média FROM Dim_Usuario INNER JOIN Dim_Situacao_Sanguinea 
    ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id GROUP by Profissão
    
'''

lista_imc_profissao = pd.read_sql_query(query1, con=dbconn)
lista_imc_profissao
```

## 3 - Total dos Imcs por Cidade

```python
query2 = '''
    
    SELECT Cidade, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média FROM Dim_Endereco INNER JOIN Dim_Situacao_Sanguinea 
    ON Dim_Situacao_Sanguinea.id = Dim_Endereco.id GROUP by Cidade
    
'''

lista_imc_endereco = pd.read_sql_query(query2, con=dbconn)
lista_imc_endereco
```

## 4 - Total dos Imcs por Tipo Sanguineo

```python
query3 = '''
    
    SELECT Tipo_Sanguineo, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média FROM Dim_Situacao_Sanguinea GROUP BY Tipo_Sanguineo
    
'''

lista_imc_sanguineo = pd.read_sql_query(query3, con=dbconn)
lista_imc_sanguineo
```

## 5 - Total dos Imcs por Sexo

```python
query4 = '''
    
    SELECT Sexo, Nome_Completo, SUM(Imc) AS Total FROM Dim_Usuario INNER JOIN Dim_Situacao_Sanguinea 
    ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id GROUP BY Nome_Completo, Sexo
    
'''

lista_imc_sexo = pd.read_sql_query(query4, con=dbconn)
lista_imc_sexo
```

## 6 - Mostre na coluna Ultima Consulta as datas referentes aos anos acima de 2018 

```python
query5 = '''
    
    SELECT  Nome_Completo, Última_Consulta FROM Dim_Usuario INNER JOIN 
    Dim_Situacao_Sanguinea ON 
    Dim_Situacao_Sanguinea.id = Dim_Usuario.id WHERE Última_Consulta > '2018-01-21 00:00:00' 
    ORDER BY Última_Consulta
    
'''

lista_ultima_consulta = pd.read_sql_query(query5, con=dbconn)
lista_ultima_consulta
```

## 7 - Total dos Imcs por Dia, Ano e Mês

```python
query6 = '''
    
    SELECT Dia_Consulta, Ano_Consulta, Mês_Consulta, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média FROM Dim_Situacao_Sanguinea 
    GROUP BY Ano_Consulta
    
'''

total_dia_ano_mes = pd.read_sql_query(query6, con=dbconn)
total_dia_ano_mes
```

## 8 - Faça uma consulta que mostre o Imc dos nomes dos usuarios com as respectivas cidades e o Tipo Sanguíneo

```python
query7 = '''
    
   SELECT Nome_Completo, Tipo_Sanguineo, Cidade, SUM(Imc) AS Total, ROUND(AVG(Imc),2) AS Média
   FROM Dim_Usuario 
   INNER JOIN Dim_Situacao_Sanguinea ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id 
   INNER JOIN Dim_Endereco ON Dim_Endereco.id = Dim_Usuario.id GROUP BY Cidade
'''

consulta_nomes_cidades_ts_imcs = pd.read_sql_query(query7, con=dbconn)
consulta_nomes_cidades_ts_imcs
```

## 9 - Quais são as pessoas com o tipo sanguineo que está acima e abaíxo do peso

** Acima do Peso

```python
acima_peso = '''
    
  SELECT Nome_Completo, Tipo_Sanguineo, Imc FROM Dim_Usuario 
  INNER JOIN Dim_Situacao_Sanguinea ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id WHERE Imc >= 25
'''

analise_acima = pd.read_sql_query(acima_peso, con=dbconn)

print('acima do peso') 
display(analise_acima)
```

** Abaixo do Peso


```python
abaixo_peso = '''
    
  SELECT Nome_Completo, Tipo_Sanguineo, Imc FROM Dim_Usuario 
  INNER JOIN Dim_Situacao_Sanguinea ON Dim_Situacao_Sanguinea.id = Dim_Usuario.id WHERE Imc <= 25
  
'''

analise_abaixo = pd.read_sql_query(abaixo_peso, con=dbconn)

print('acima do peso') 
display(analise_abaixo)
```
